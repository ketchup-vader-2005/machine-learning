```python
import tensorflow as tf
import tf2onnx

# Reload the model
model = tf.keras.models.load_model("CNNmodel.keras")

# Run export code
spec = (tf.TensorSpec((None, 321, 6), tf.float32, name="feature_input"),)
tf2onnx.convert.from_keras(model, input_signature=spec, opset=13, output_path="test_model.onnx")
```
